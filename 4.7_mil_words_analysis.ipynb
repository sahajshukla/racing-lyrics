```python
count = df.groupby(['year']).agg(['sum'])
```

```python
count = pd.DataFrame(count)
count
```

```python
count['ratio'] = count['year_total']/count['year_count']
```

```python
counter = count['ratio']
```

```python
counter = pd.DataFrame(counter)
```

```python
counter.columns
```

```python
counter.ratio = counter.ratio.astype(int)
```

```python
counter
```

```python
new = []
for i in counter.ratio:
    for j in range(i):
        new.append(i)
```

```python
new = pd.DataFrame(new)
new 
```

```python
df = df.join(new)
df.columns
df.rename({0:'distinct count'}, axis = 1, inplace = True)
df
```

```python
df['average_count'] = df['year_total']/df['distinct count']

```

```python
df
```

```python
df['word_weightage'] = df['per_mil']/df['average_count']
english_stopwords = sw.words('english')
french_stopwords = sw.words('french')
spanish_stopwords = sw.words('spanish')
```

```python
new_var = []
for row in df.iloc[:,0]:
    if row in english_stopwords or row in french_stopwords or row in spanish_stopwords:
        df = df[df['word']!=row]
```

```python
group_data = df.groupby(['year']).apply(lambda x : x.sort_values(by = ['word_weightage'], ascending = False).head(30))
```

```python
# One can clearly observe that the language has been more generic lately instead of being specific to a word itself. The lyrics have seemingly become more casual due to ease in availibility and a wider acceptance in general
# this is also indicative of the change in the language itself over the century 
```

```python
group_data = group_data.rename(columns={'year':'yeaar'})
```

```python
group_data = pd.DataFrame(group_data)
group_data['cumsum'] = group_data.groupby(['word'])['word_weightage'].cumsum(axis=0)
#cumsum is used to handle the unbalanced part of the dataset. We first take the expected number of times a word should repeat and then compare it to actual repeats
```
